```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

# Load the Audio

```python
audio_path = "../AUDIOS_DIR/-0SdAVK79lg.wav"
```

What is a Time-Domain Signal?

- A time-domain signal represents the amplitude of the audio signal over time.

- It is typically a 1D array where each value corresponds to the amplitude of the audio at a specific point in time.

```python
y, sr = librosa.load(audio_path, sr=None, mono=True)  # y: audio signal, sr: sample rate

print("Audio signal shape (y):", y.shape)
print("Sample rate (sr):", sr)
```

y: A 1D numpy array representing the audio signal.

sr: The sample rate (number of samples per second).`

# Compute the STFT (Short-Time Fourier Transform)

What is the STFT?

- The STFT divides the audio signal into overlapping frames and computes the Fourier transform for each frame.

- The result is a spectrogram, which shows how the frequency content of the signal changes over time.



```python
# STFT parameters
n_fft = 2048  # Frame size (number of samples per frame)
hop_length = 512  # Frame shift (number of samples between frames)

# Compute the STFT
stft = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)

# Convert the STFT to magnitude (absolute value)
magnitude = np.abs(stft)

# Print the shape of the STFT
print("STFT shape:", magnitude.shape)
```

stft: A 2D numpy array where:

  - Rows represent frequency bins.

  - Columns represent time frames.

magnitude: The magnitude of the STFT (absolute value of the complex STFT).

# Convert to Mel Scale

The mel scale is a perceptual scale of pitches that approximates how humans perceive sound. Converting the STFT to the mel scale involves applying a mel filter bank.

- What is a Mel Filter Bank?

    A mel filter bank is a set of triangular filters spaced according to the mel scale.

Each filter corresponds to a specific frequency range and is used to compute the energy in that range.
