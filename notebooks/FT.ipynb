````python
import librosa
import scipy as sp
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
````

````python
audio_file = '../piano-g-6200.mp3'
````

````python
signal, sr = librosa.load(audio_file)
````

````python
print(sr)
````

````python
ipd.Audio(audio_file, rate=22050)
````

````python
plt.figure(figsize=(15,8))
librosa.display.waveshow(signal, sr=sr, alpha=0.5)
plt.show()
````

````python
# getting sepctrums using fourier transformation
ft = sp.fft.fft(signal)
print(type(ft))
print(ft)
magnitude = np.absolute(ft)
print(type(magnitude))
print(len(magnitude))
frequency = np.linspace(0,sr,len(magnitude))
print(len(frequency))
````

````python
plt.figure(figsize=(18,8))
plt.plot(frequency[:5000],magnitude[:5000])
plt.xlabel("frequency")
plt.ylabel("magnitude")
plt.show()
````

````python
# zoom into that waveform
samples = range(len(signal))
t = librosa.samples_to_time(samples, sr=sr)

plt.figure(figsize=(18,8))
plt.plot(t[10000:10400], signal[10000:10400])
plt.title("Zoom into waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
````

#### Sin Wave
This is a formula for sinosidal wave -> sin(2pi(ft - phase))


````python
f = 876
phase = 1.0
# phase1 = 0.7

sin = 0.5 * np.sin(2 * np.pi * (f * t - phase))
# sin1 = 0.5 * np.sin(2 * np.pi * (f * t - phase1))

plt.figure(figsize=(18,8))
plt.plot(t[10000:10400], sin[10000:10400], color = 'r')
plt.plot(t[10000:10400], signal[10000:10400])

plt.fill_between(t[10000:10400],sin[10000:10400]*signal[10000:10400],color="g")

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
````
